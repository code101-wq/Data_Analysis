# Working with numpy 
Numpy is a python library built on c++

They are used to build matrices which are used to manipulate data

```python
mydata = np.arange(0, 20)
mydata
```

```python
np.reshape(mydata, (4,5)) # 5 rows 4 columns
```

```python
Matr1 = np.reshape(mydata, (5, 4), order='c') # 'c'(row-major) means to read / write elements in a c-like index order
Matr1 # Read/Writes elements by rows(last index changes fastest)
```

```python
Matr2 = np.reshape(mydata, (5,4), order='F') # 'F' (column-major) Read/Writes elements in a fortran index order
Matr2 # Read/Writes elements by columns (first index changes fastest) 
```

```python
Matr3 = np.reshape(mydata, (5,4), order='A') # (Allowed) Read/Writes elements depending on the memory layourt
Matr3 # If column order is used it would behave like F, else it wioll be have like C
```
